```python
#Exponential smoothing for forecasting -
# https://www.youtube.com/watch?v=Wf2naBHRo8Q 
# https://www.youtube.com/watch?v=GbRfbXHXUKM

#Importing necessary libraries 
import pandas as pd
from datetime import datetime
import numpy as np
import sktime
import matplotlib.pyplot as plt 
from sktime.forecasting.exp_smoothing import ExponentialSmoothing
from sktime.forecasting.base import ForecastingHorizon
from sktime.utils.plotting import plot_series
from sktime.forecasting.model_selection import temporal_train_test_split 

%matplotlib inline
```

```python
df = pd.read_json("./mergedJson.json")

# Cleaning the data 

# Checking the negative values - possible time out ???
df_test = df[df['avg']<0]
df_test.groupby(['prb_id','dst_addr']).count()

# Change them to NaN
df['avg'].replace(-1.0, np.nan, inplace=True)

# Checking for null values in each column
df_null = df['avg'].isnull().sum()
df_null

# Remove Null values 
df = df.dropna(how='any',axis=0) 
df.shape
```

```python
#Changing the time column from epoch to date time format for time series processing 
new_timestamp = []

for i in df['stored_timestamp']:
    my_datetime = datetime.fromtimestamp(i)
    new_timestamp.append(my_datetime)
    
#Renaming the probe_id column for easy plotting 
nprb_id = []

for value in df["prb_id"]:
    if value == 1004991:
        nprb_id.append('es1')
    if value == 53229:
        nprb_id.append('es2')
    if value == 1004997:
        nprb_id.append("pt1")
    if value == 1004102:
        nprb_id.append("pt2")
    if value == 20757:
        nprb_id.append("es3")
    if value ==1003454:
        nprb_id.append("nl1")
    if value == 1003158:
        nprb_id.append("nl2")
    if value == 1003747:
        nprb_id.append("nl3")
    if value == 1004200:
        nprb_id.append("es4")
    if value == 51381:
        nprb_id.append("ug1")
    if value == 54470:
        nprb_id.append("us1")
    if value == 1002914:
        nprb_id.append("us2")
        
```

```python
#preparing the data we will use for forecasting 

df2 = df.copy()

df2['new_time'] = new_timestamp
df2["nprb_id"] = nprb_id
df2.groupby(['nprb_id','dst_addr'])

#Extracting the results of one probe and destination
dfprbdtn = df2[(df2['nprb_id'] == 'es1') & (df2['dst_addr'] == "52.46.200.93")]
df_sktime = dfprbdtn.set_index(dfprbdtn['new_time'])[["avg"]]
df_sktime
```

```python
#adding fixed temporal distance to data.Required by library 
df_sktime = df_sktime.asfreq('1S', method='ffill')
df_sktime = df_sktime.asfreq('300S')
df_sktime.index
```

```python
#looking at general trend of data 
train , test = temporal_train_test_split(df_sktime)
plot_series(train,test)
```

```python
train
```

```python
test
```

```python
#choosing the model and training it 
es_model = ExponentialSmoothing(smoothing_level=0.001, optimized=False)
es_model = es_model.fit(train)
```

```python
#add a forecasting horizon 
fh = ForecastingHorizon(test.index,is_relative=False)
fh
```

```python
#Generating prediction
es_forecast = es_model.predict(fh)
es_forecast
```

```python
# Comparing test and predicted values 
test.rename(columns = {'avg':'test_avg' }, inplace = True)
es_forecast.rename(columns = {'avg':'forecast_avg' }, inplace = True)
result_df = pd.concat([test.reset_index(drop=True), es_forecast.reset_index(drop=True)], axis=1)
result_df
```

```python
# Testing the prediction accuracy 
se = (result_df['forecast_avg'] - result_df['test_avg']) ** 2 
se.mean()
```

```python
plot_series(train,test,es_forecast)
```

```python
#checking other destinations for same probe 
probes = ['es1','es2','es3','es4','nl1','nl3','pt2','ug1','us1','us2']
dstn = ['52.46.72.50','52.94.56.52','54.239.55.167','67.220.226.67']
df_pdn = []

for probe in probes:
    for dst in dstn:
    
        dfpdn = df2[(df2['nprb_id'] == probe) & (df2['dst_addr'] == dst)]
        df_pdn.append(dfpdn)
        
        df_sktime = dfpdn.set_index(dfpdn['new_time'])[["avg"]]
        df_sktime = df_sktime.asfreq('1S', method='ffill')
        df_sktime = df_sktime.asfreq('300S')

        train , test = temporal_train_test_split(df_sktime)
        
        es_model = ExponentialSmoothing(smoothing_level=0.001, optimized=False)
        es_model = es_model.fit(train)
        
        fh = ForecastingHorizon(test.index,is_relative=False)
        es_forecast = es_model.predict(fh)
        plot_series(train,test,es_forecast)
   
        
        
```
